```
import lambdata_unburied.modints as mi
```

```
import seaborn as sns
import pandas as pd

data = sns.load_dataset('iris')

data.head()
```

```
from sklearn.ensemble import RandomForestClassifier

target = 'species'
features = data.columns.tolist()
features.remove(target)

X = data[features]
y = data[target]

model = RandomForestClassifier()
model.fit(X,y)
```

```
%matplotlib inline

mi.permuter(model, X, y)
```

```
mi.isolated(model, X, features[0])
```

```
mi.interaction(model, X, [features[0], features[1]])
```
